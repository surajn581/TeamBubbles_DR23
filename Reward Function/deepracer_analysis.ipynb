```python
def calculate_angle(x1, y1, x2, y2, x3, y3):
    # Calculate the angles between two line segments formed by (x1, y1), (x2, y2), and (x3, y3)
    dx1 = x2 - x1
    dy1 = y2 - y1
    dx2 = x3 - x2
    dy2 = y3 - y2

    angle1 = math.atan2(dy1, dx1)
    angle2 = math.atan2(dy2, dx2)

    angle_degrees = math.degrees(angle2 - angle1)
    
    if angle_degrees < 0:
        angle_degrees += 360

    return angle_degrees

def shape_complexity(coordinates):
    num_angles = len(coordinates) - 2
    total_angle_sum = 0

    for i in range(num_angles):
        x1, y1 = coordinates[i]
        x2, y2 = coordinates[i + 1]
        x3, y3 = coordinates[i + 2]

        angle = calculate_angle(x1, y1, x2, y2, x3, y3)
        total_angle_sum += angle

    # Calculate the average angle to assess shape complexity
    average_angle = total_angle_sum / num_angles

    # Higher average angle indicates a more complex shape
    return average_angle
```

```python
# Generate points along the circumference of a circle
def generate_circle_points(center, radius, num_points):
    x0, y0 = center
    points = []
    for i in range(num_points):
        angle = 2 * math.pi * i / num_points
        x = x0 + radius * math.cos(angle)
        y = y0 + radius * math.sin(angle)
        points.append((x, y))
    return points
```

```python
def is_a_turn_coming_up( wp, n_points, angle_threshold ):
    angles = [ angle( wp[i], wp[i+1] ) for i in range( min( n_points-1, len(wp)-1 ) ) ]
    angles = [ normalize_angle_to_360(angle) for angle in angles ]
    diff_angles = [ abs(angles[i] - angles[i+1]) for i in range(len(angles) - 1) ]
    if sum(diff_angles) > angle_threshold:
        return True
    return False
    #return not all([ diff < angle_threshold for diff in diff_angles ])
```

```python
def calculate_angle(p1, p2, p3):
    # Calculate the angle between three points p1, p2, p3
    angle = math.degrees(
        math.atan2(p3[1] - p2[1], p3[0] - p2[0]) - math.atan2(p1[1] - p2[1], p1[0] - p2[0])
    )
    return angle % 360

def get_turn_points(coordinates, n_points = 8, angle_threshold = 3):
    turn_points = []
    window_size = n_points
    threshold_angle = angle_threshold  # Set a threshold angle to determine a significant turn

    for i in range(len(coordinates) - window_size + 1):
        window = coordinates[i : i + window_size]
        angles = [
            calculate_angle(window[j], window[j + 1], window[j + 2]) for j in range(window_size - 2)
        ]
        max_angle_change = abs( max(angles) - min(angles) )
        if max_angle_change >= threshold_angle:
            turn_points.append(coordinates[i])
    return turn_points

# Example usage:
#race_track_coordinates = [(x, y) for x, y in zip(range(100), range(100))]  # Replace with your actual coordinates
turn_points = get_turn_points(coordinates, 5, 4.9)
print("Is part of a turn:", turn_points)
```
